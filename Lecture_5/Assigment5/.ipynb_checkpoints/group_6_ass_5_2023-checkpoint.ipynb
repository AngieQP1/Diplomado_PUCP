```python
import pandas as pd
import calendar

def is_weekend(year, month, day):
    # Check if the given date is a Saturday or Sunday
    return calendar.weekday(year, month, day) in [calendar.SATURDAY, calendar.SUNDAY]

def import_data_sbs_banks(start_month, start_year, end_month=None, end_year=None):
    # Check if start_month and start_year are provided
    if start_month is None or start_year is None:
        raise ValueError("Error: Start month and start year must be provided.")
    
    # Initialize the nested dictionary to store DataFrames
    data_dict = {}

    # Get the current year and month
    current_year, current_month = pd.Timestamp.now().year, pd.Timestamp.now().month

    # Determine the end_year value
    if end_year is None:
        end_year_value = current_year
    else:
        end_year_value = end_year

    # Iterate over the range of years and months
    for year in range(start_year, end_year_value + 1):
        for month in range(start_month, end_month + 1) if year == start_year and year == end_year_value else range(1, 13):
            # Skip months after the end month if end_month and end_year are provided
            if end_month is not None and end_year is not None and (year == end_year and month > end_month):
                continue

            # Skip future months if end_month and end_year are not provided
            if end_month is None and end_year is None and (year == current_year and month > current_month):
                continue

            # Determine the last day of the month
            last_day = 31 if month in [1, 3, 5, 7, 8, 10, 12] else 30 if month == 4 or month == 6 or month == 9 or month == 11 else 28

            # Check if the last day is a Saturday or Sunday
            if is_weekend(year, month, last_day):
                # If the last day is a weekend, consider the previous day as the last day
                last_day -= 1
                if is_weekend(year, month, last_day):
                    # If the previous day is also a weekend, consider the day before as the last day
                    last_day -= 1

            # Construct the file name based on the current year, month, and the last day
            file_name = f'table_clean_{last_day}_{month}_{year}.xlsx'

            # Specify the path to the Excel file
            excel_path = fr'..\..\_data\sbs\B_RawData\bancos\{file_name}'

            try:
                # Read the Excel file into a pandas DataFrame
                data_frame = pd.read_excel(excel_path)

                # Create a nested dictionary entry
                if year not in data_dict:
                    data_dict[year] = {}
                data_dict[year][month] = data_frame

            except FileNotFoundError:
                # Continue with the loop if the file is not found without printing an error message
                continue

    return data_dict

import_data_sbs_banks(2,2004,3,2004)
```
