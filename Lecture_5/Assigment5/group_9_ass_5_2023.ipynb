#### 1.3 Generación de la función "import_data_sbs_banks"

```python
#Stage 1: Dictionary Initialization

# For this stage, we describe a function called import_data_sbs_banks. The function aims to import data
# from a specified folder and store them in two dictionaries: "dict_banks" for the final data and "dict_years"
# for data for each year.

# In the first line of the function, the dict_banks dictionary is initialized as an empty dictionary to be used
# to store the final data. In the next line, the dict_years dictionary is also initialized as a
# empty dictionary, which will be used to store the data broken down by year.

# Additionally, the function is defined with start parameters (start_month and start_year) and end parameters with
# default values (end_month and end_year). This allows the function to work even if no
# values for the final month and year, defaulting to August and 2022 respectively. These parameters
# flexibles provide versatility to the code by allowing you to adjust the range of data to import according to the needs you
# are required.
def import_data_sbs_banks(start_month , start_year , end_month = 8, end_year = 2022):
    dict_banks = {}                 
    dict_years = {}  

# Stage 2: Loop for years
    #We used a loop that will iterate over each year contained in the range. The loop will start on start_year and end on end_year + 1
    for anio in range(start_year, end_year + 1):    # We defined the parameters for the loop based in the arguments of import_data_sbs_banks()
    
# Stage 3: Loop for months
        #Within the loop for years, we created another loop that will iterate over each month in the range: from start_month to end_month +1.
        for mes in range(start_month, end_month + 1):   # We defined the parameters for the loop based in the arguments of import_data_sbs_banks()
            
# Stage 4: Loop for days:
# Within the loop for months, we added a third loop that will iterate over the 26th o the 31st days of the month
            for dia in range(26,32):

# Stage 5: Attempt to read Excel files
#In this step, an attempt is made to read Excel files using the pandas library. 
#The 'pd.read_excel' function is used to load an excel file into a pandas dataFrame. 
#The file name is constructed using information from the loops over years, months, and days.
# 'str(dia)': Represents the current day in the inner day loop.
# 'str(mes)': Represents the current month in the inner month loop.
# 'str(anio)': Represents the current year in the outer year loop.
#The excel file is expected to have the name "table_clean_day_month_year.xlsx". 
#If the reading is successful, a dataframe named 'df' is created with the data from the excel file. 
#Then, this dataframe is stored in a variable with a dynamic name based on the current month and year from the loop using 'locals()[f'df{mes}_{anio}'].
#The 'try' and 'except' block is used to handle situations where the file does not exist or there is an issue in reading the file. 
#If an error occurs during the file reading, it is skipped, and the loop continues to the next day.
                try:
                    df = pd.read_excel(r"../../_data\sbs\B_RawData\bancos\table_clean_"+str(dia)+"_"+str(mes)+"_"+str(anio)+".xlsx")
                    locals()[f'df{mes}_{anio}'] = df    
                except:
                    continue

# Stage 6: Handling dictionaries
#In this step, two dictionaries are managed: 'dict_banks' and 'dict_years'. 
#These dictionaries are used to store the dataframes generated in the previous step.
# Handling 'dict_years': If the current month ('mes') is not present in the 'dict_years' dictionary, an entry is created for that month. 
#                      Then, the dataframe generated in the previous step is assigned to this entry. 
#                      If the month is already present, nothing is done, and the loop continues to the next day.
# Handling 'dict_banks': Similar to the handling of 'dict_years', it is checked whether the current year (anio) is present in the 'dict_banks' dictionary.
#                      If it is not present, an entry is created for that year. 
#                      Then, the 'dict_years' dictionary (containing dataframes organized by month) is assigned to this entry. 
#                      If the year is already present, nothing is done, and the loop continues to the next year.
#At the end of these steps, the function constructs and returns a nested dictionary (dict_banks) containing DataFrames organized by year, month, and day. This dictionary is the final result of the function.
                if mes not in dict_years:
                    dict_years[mes] = {}
                else:
                    continue
                dict_years[mes]=locals()[f'df{mes}_{anio}']
        if anio not in dict_banks:
            dict_banks[anio] = {}
        else:
            continue
        dict_banks[anio]=dict_years
        
# Stage 7: Final result
# The previous six steps execution allows us to obtain as a result a nested dictionary (dict_banks) that contains the Dataframes
# organized by year and month
    return dict_banks
```
