```python
# Nombres deseados para los archivos y etiquetas
file_names = ['rec_1', 'rec_2', 'rec_3']
var_labels_names = ['var_labels1', 'var_labels2', 'var_labels3']
value_labels_names = ['value_labels1', 'value_labels2', 'value_labels3']
# Nombres deseados para los archivos y etiquetas
file_names = ['rec_1', 'rec_2', 'rec_3']
var_labels_names = ['var_labels1', 'var_labels2', 'var_labels3']
value_labels_names = ['value_labels1', 'value_labels2', 'value_labels3']

# Diccionarios para almacenar datos y etiquetas
data_dict = {}
var_labels_dict = {}
value_labels_dict = {}

# Cargar archivos y asignar nombres
for i, file_path in enumerate(file_paths):
    data, meta = pyreadstat.read_sav(file_path)
    
    # Asignar nombres de archivos
    data_dict[file_names[i]] = data
    
    # Asignar nombres de etiquetas de variables
    var_labels_dict[var_labels_names[i]] = meta.column_labels
    
    # Asignar nombres de etiquetas de valores
    value_labels_dict[value_labels_names[i]] = meta.variable_value_labels

# Acceder a los datos y etiquetas seg√∫n los nombres especificados
rec_1_data = data_dict['rec_1']
var_labels1 = var_labels_dict['var_labels1']
value_labels1 = value_labels_dict['value_labels1']

rec_2_data = data_dict['rec_2']
var_labels2 = var_labels_dict['var_labels2']
value_labels2 = value_labels_dict['value_labels2']

rec_3_data = data_dict['rec_3']
var_labels3 = var_labels_dict['var_labels3']
value_labels3 = value_labels_dict['value_labels3']
```
