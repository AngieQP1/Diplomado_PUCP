The name of imported files should be named as `rec_1`, `rec_2` and `rec_3` for files `REC0111.sav`, `RE223132.sav` and `RE516171.sav` respectively. The name of the variable and value labels should be `var_labels1` and `value_labels1` for `rec1`, `var_labels2` and `value_labels2` for `rec2`, and `var_labels3` and `value_labels3` for `rec3`. **Hint: See the section 3.3.4 of [the lecture 3](https://github.com/alexanderquispe/Diplomado_PUCP/blob/main/Lecture_3/Lecture_3.ipynb)**

```python


# List of SPSS file paths
file_paths = [
    r"../../_data/endes/2019/REC0111.sav",
    r"../../_data/endes/2019/RE223132.sav",
    r"../../_data/endes/2019/RE516171.sav",
]

# Lists to store DataFrames and metadata dictionaries
rec_dfs = []
var_labels_list = []
value_labels_list = []

# Loop through the file paths
for i, file_path in enumerate(file_paths):
    # Read SPSS file
    rec_df = pd.read_spss(file_path)
    rec_dfs.append(rec_df)

    # Obtain metadata
    with sav.SavHeaderReader(file_path, ioUtf8=True) as header:
        metadata = header.all()

    # Save variable labels and value labels
    value_labels = metadata.valueLabels
    var_labels = metadata.varLabels
    
    # Adding labels from sav file    
    rec_df.attrs[ 'var_labels' ] = var_labels
    rec_df.attrs[ 'value_labels' ] = value_labels
    # Assign a name to the DataFrame based on the index
    rec_df_name = f"rec_{i + 1}"
    rec_var_name = f"var_labels{i + 1}"
    rec_val_name = f"value_labels{i + 1}"
    globals()[rec_df_name] = rec_df
    globals()[rec_var_name] = var_labels
    globals()[rec_val_name] = value_labels

# Now, rec_dfs contains the DataFrames, var_labels_list contains variable labels, and value_labels_list contains value labels.
# Additionally, rec_1, rec_2, and rec_3 are assigned to the DataFrames.
```

## 7. Mean key vars

7. Use `endes_2019` data to generate a new object named `mean_key_vars` to find the mean of **total children ever born (V201)**, **Ideal number of children (V613)**, **Husbands education-single yrs (V715)**, and **Age at first marriage (V511)** by year and department **(V024)**. Name these columns as **mean_total_children, mean_ideal_children, mean_hb_yr_educ and mean_first_marriage**, respectively. **Hint: Use groupby and [this link](https://stackoverflow.com/questions/40901770/is-there-a-simple-way-to-change-a-column-of-yes-no-to-1-0-in-a-pandas-dataframe).**

```python
# #Select the columns to group by
group_columns = ['year', 'V024']

# Select the vars
key_vars = ["V201","V613","V715","V511"]
#df['column_name'] = pd.factorize(df['column_name'])[0]

# Calculate mean by department
mean_key_vars = endes_2019.groupby(group_columns)[key_vars].mean()


# Reset the index to obtain a dataFrame
mean_key_vars.reset_index(inplace=True)

# Rename in order the variables
mean_key_vars.columns = ['year', 'dpto','mean_total_children', 'mean_ideal_children', 'mean_hb_yr_educ', 'mean_first_marriage']
```
